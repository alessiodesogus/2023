## 2.1 Visualizing uncertainty

- How did the average revenue of blockbusters evolve? Visualize the average yearly worldwide gross revenue, together with the standard deviation. Below, you are provided with two rough sketches of how the visualization could look like. We could indicate the margin of error with lines with caps, or by contouring between the upper and lower bound.

- Hints: use Matplotlib errorbar and fill_between functions.

- Did the average revenue increase/decrease over time? What do you observe?


<img src="figure0.jpg" style="width: 400px;" align="left" />
<img src="figure1.jpg" style="width: 400px;" align="left"/>


<br>

```python
# Calculate yearly stats across movies here
# Group data by year
yearly_stats = movies.groupby('year')

# Calculate the average and standard deviation of worldwide gross revenue for each year
total_worldwide_gross = yearly_stats['worldwide_gross'].sum()
average_gross = yearly_stats['worldwide_gross'].mean()
std_deviation = yearly_stats['worldwide_gross'].std()

# New Data Frame
yearly_stats_df = pd.DataFrame({
    'Total_Worldwide_Gross': total_worldwide_gross,
    'Average_Grossing_Movie': average_gross,
    'Standard_Deviation': std_deviation
}).reset_index()

# Extract years for the x-axis
years = yearly_stats_df['year']

# Display
yearly_stats_df.head(5)
```

```python
# Write your code to make plot with errorbars here
# Create the figure and axis
fig, ax = plt.subplots(figsize=(8, 4))

# Plot the average revenue as a line with markers
ax.errorbar(years, average_gross, yerr=std_deviation, fmt='o-', label='Average Gross Revenue with Std Deviation')

# Add labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Worldwide Gross Revenue')
ax.set_title('Average Yearly Worldwide Gross Revenue with Standard Deviation')

# Add a legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Write your code to make filled plot here
# Create the figure and axis
fig, ax = plt.subplots(figsize=(8, 4))

# Plot the average revenue as a line
ax.plot(years, average_gross, marker='o', linestyle='-', label='Average Gross Revenue')

# Fill between upper and lower bounds to indicate the margin of error
ax.fill_between(years, average_gross - std_deviation, average_gross + std_deviation, color='gray', alpha=0.4, label='Std Deviation')

# Add labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Worldwide Gross Revenue')
ax.set_title('Average Yearly Worldwide Gross Revenue with Standard Deviation (Filled Plot)')

# Add a legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

## 2.2 Creating panels

- Next, we want to understand how long the blockbusters are in the different genres. When working with a large number of categories (we have 16 genres in the movies dataset), the visualizations easily become cluttered, and it becomes useful to make panels with small subplots.

- Make a 4x4 panel for the 16 genres. In each subplot, show a histogram with the distribution of movie lengths for different categories. Below, find a skets of how this visualization could look like. Note the same x- and y-scale that allows comparison between categories!

- Hints: use Matplotlib subplots function to initialize the subplotting axes.

- What genres do the longest movies belong too? What can you say about the length of Comedy and animation movies?

- How could we assign different color to each subplot (each genre)? What type of palette would be the best choice? Make a variation of the visualzation where each genre is visualized in a different color.

<img src="figure3.jpg" style="width: 500px;" align="left">

```python
# Calculate stats by genre here
# Group data by genre
genre_stats = movies.groupby('Main_Genre')

# Calculate the overall minimum and maximum movie length for consistent scaling
min_length = movies['length'].min()
max_length = movies['length'].max()
```

```python
# Write your code to make 4x4 panel here for the 16 genres
# We use the sharex and sharey parameters when creating the subplots to ensure that all subplots share the same x-axis and y-axis.
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(16, 16), sharex=True, sharey=True)

# Iterate through genres and create histograms
for i, (genre, data) in enumerate(genre_stats):
    row = i // 4  # Determine the row of the subplot
    col = i % 4   # Determine the column of the subplot
    
    # Create a histogram for the current genre
    axes[row, col].hist(data['length'], bins=20, range=(min_length, max_length), alpha=0.7, color='skyblue')
    
    # Set titles and labels
    axes[row, col].set_title(genre)
    axes[row, col].set_xlabel('Movie Length (minutes)')
    
    # Set y-axis to represent the number of movies
    axes[row, col].set_ylabel('Number of Movies')

# Show the panel of histograms
plt.tight_layout()
plt.show()
```

```python
# Make a variation of the visualzation where each genre is visualized in a different color.
# We use the sharex and sharey parameters when creating the subplots to ensure that all subplots share the same x-axis and y-axis.
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(16, 16), sharex=True, sharey=True)

# Define a color palette (using Seaborn's "Set1" palette)
color_palette = sns.color_palette("Set1", n_colors=len(genre_stats))

# Iterate through genres and create histograms with different colors
for i, ((genre, data), color) in enumerate(zip(genre_stats, color_palette)): 
    row = i // 4  # Determine the row of the subplot
    col = i % 4   # Determine the column of the subplot
    
    # Create a histogram for the current genre with a different color
    axes[row, col].hist(data['length'], bins=20, range=(min_length, max_length), alpha=0.7, color=color)
    
    # Set titles and labels
    axes[row, col].set_title(genre)
    axes[row, col].set_xlabel('Movie Length (minutes)')
    
    # Set y-axis to represent the number of movies
    axes[row, col].set_ylabel('Number of Movies')

# Show the panel of histograms
plt.tight_layout()
plt.show()

# We iterate through the genres and assign a different color to each genre by using the zip function to combine genre data and colors.
```

## 2.3 Heatmaps

- What studios make blockbusters in which genre? 

- There are many studios and many genres. When working with two variables that both have many different categories, there is often no space for representing each combination of two categories with a detailed plot (such as barplot of boxplot). It becomes useful to represent each stratum with one number, indicated with color in a heatmap. Such visualizations are useful to quickly understand the data.

- Visualize the number of blockbusters made by each studio, of each genre (first heatmap). You will find below a rough sketch. Hint: use Seaborn heatmap function. 

- Finally, with a heatmap we can represent three variables, by coding a third variable with color. Try to understand what is the relationship between genres by understanding the co-occurence patterns between the main and secondary genre. What is the average worldwide gross revenue for movies with different combinations of main and second genre? (second heatmap)

- You should be able to easily answer the following questions from the two visualizations: What genres do the majority of movies made by Warner Bros studio fall into? What is the most successful genre of Walt Disney pictures movies? What studio made the largest number of Romance blockbusters? Which combination of genres brings the highest revenue?



<img src="figure4.jpg" style="width: 500px;" align="left">

```python
# Write your code to make the first heatmap here
# Pivot the data to create a matrix of counts for each combination of studio and genre
heatmap_data = movies.pivot_table(index='studio', columns='Main_Genre', aggfunc='size', fill_value=0)

# Create a heatmap
# annot=True is for display the number in each case
# There are many different color maps : 'viridis', 'coolwarm', 'cividis', 'coolwarm', 'plasma', 'YlOrRd', 'BuGn'
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="coolwarm", annot=True, fmt="d", linewidths=0.5) 

# Add labels and a title
plt.xlabel("Main Genre")
plt.ylabel("Studio")
plt.title("Number of Blockbusters by Studio and Genre")

# Show the heatmap
plt.show()


```

```python
# Write your code to make the second heatmap here - Another way of doing things with a "heatmap_data" here :

# Pivot the data to create a matrix of the mean of worldwide gross revenue for each combination of studio and genre
heatmap_data = movies.pivot_table(index='studio', columns=['Main_Genre', 'Genre_2'], values='worldwide_gross', aggfunc='mean', fill_value=0)

# Create a heatmap with the 'viridis' colormap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="coolwarm", annot=False, fmt=".0f", linewidths=0.5)

# Add labels and a title
plt.xlabel("Genre (Main_Genre, Genre_2)")
plt.ylabel("Studio")
plt.title("Mean Worldwide Gross Revenue by Studio and Genre")

# Show the heatmap
plt.show()

# I'm not sure about this one, there is too much column ? 
```
